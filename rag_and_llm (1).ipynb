```python
import os
from langchain.document_loaders import UnstructuredPDFLoader
from langchain.text_splitter import CharacterTextSplitter
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_chroma import Chroma
from langchain_groq import  ChatGroq
from langchain.chains import RetrievalQA
```

```python
os.environ["GROQ_API_KEY"]="use_your_grok_api key"
```

```python
#fetch the pdf from url
import requests
url = "https://pythonclassroomdiary.wordpress.com/wp-content/uploads/2019/06/functions-inpython.pdf"
response = requests.get(url)
```

```python
response
```

```python
#save the pdf in local file
#wb= writ binary
with open("functions-inpython.pdf","wb") as f:
  f.write(response.content)
```

```python
#load the document
from langchain_community.document_loaders import UnstructuredFileIOLoader

with open("functions-inpython.pdf", "rb") as f:
  loader = UnstructuredFileIOLoader(file=f)
  documents = loader.load()

documents
```

```python
text_splitter = CharacterTextSplitter(chunk_size= 100,chunk_overlap=50)
```

```python
text_splitter
```

```python
texts = text_splitter.split_documents(documents)
```

```python
type(texts)
```

```python
texts[2]
```

```python
embeddings = HuggingFaceEmbeddings()
```

```python
persist_directory ="vector_db"
```

```python
vectordb = Chroma.from_documents(texts, embeddings, persist_directory=persist_directory)
```

```python
print(vectordb)
```

```python
#retrival
retriever = vectordb.as_retriever()
```

```python
llm = ChatGroq(model="llama-3.3-70b-versatile",temperature=0)
```

```python
qa_chain= RetrievalQA.from_chain_type(llm=llm,chain_type="stuff",retriever=retriever,return_source_documents=True)
```

```python
query='explain about filter,map,reduce'
response = qa_chain.invoke({"query":query})
```

```python
print(response)
```
